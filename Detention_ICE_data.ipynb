```python
# Ensure 'Stay Book In Date Time' and 'Stay Book Out Date Time' are in datetime format
# We haven't done anything with this data yet, but we're going to play with the time/duration factors
# so this will help us going forward.

df_detentions['Stay Book In Date Time'] = pd.to_datetime(df_detentions['Stay Book In Date Time'])
df_detentions['Stay Book Out Date Time'] = pd.to_datetime(df_detentions['Stay Book Out Date Time'])
```

```python
# Again, before we get started, I'm still thinking about the efficacy of the data. We know that there are
# numerous NaT values in the 'Stay Book Out Date Time' field. And, prior to running this code, my thinking
# is that the NaT values mean a person is STILL in custody at a given facility.
# So, to test this, I'm going to isolate the Tacoma facility and see how many NaT values there are
# in 'Stay Book Out Time.' I think it should be a managably low number compaed with the 1,323,993 lines.

# This code will isolate Tacoma facility detentions.

tacoma_detentions = df_detentions[df_detentions['Detention Facility Code'] == 'CSCNWWA']

# This code below will count the number of NaT values in the 'Stay Book Out Date Time' field.

tacoma_detentions['Detention Book Out Date Time'].isnull().sum()

# This amounts to 1,448. vvvv
```

```python
# Define the data release date
data_release_date = pd.to_datetime('2025-07-30')

# Fill NaT values in 'Stay Book Out Date Time' with the data release date
df_detentions['Stay Book Out Date Time'] = df_detentions['Stay Book Out Date Time'].fillna(data_release_date)

# Calculate 'Stay Length' after filling NaT values
df_detentions['Stay Length'] = df_detentions['Stay Book Out Date Time'] - df_detentions['Stay Book In Date Time']
```

```python
# Create the 'Booked year' column with the year from 'Stay Book In Date Time'
df_detentions['Booked year'] = df_detentions['Stay Book In Date Time'].dt.year
```

```python
# OK, right now, df_detentions is full of multiple rows that involve a single person's single stay. I.e.,
# a person is detained, then transferred, etc. That makes multiple rows, but it's still a single stay.
# So we're looking to see if we can isolate unique stays ("unique_bookings")
# The way we do that is by grouping unique identifiers AND stay book in date times and then dropping duplicates.
# So in rows where user X014014 has 2024-01-01 1:00:00 stay, that should only appear once rather than 2-3 times.

# Find the number of unique bookings by unique identifier and stay book in date time
unique_bookings = df_detentions[['Unique Identifier', 'Stay Book In Date Time']].drop_duplicates()

# Display the number of unique bookings
print(f"The number of unique bookings is: {len(unique_bookings)}")
```

```python
# Ok, this one is important.
# We're going to lump in multiple values under 'Stay Release Reason' that seem to be times when immigration agents released a person of their own recognizance.
# The theory being that more people were released under Biden than Trump.

release_reasons_to_filter = ['order of recognizance', 'order of supervision', 'order of supervision - re-release']

filtered_detentions_df = df_detentions[
    df_detentions['Stay Release Reason'].str.lower().isin(release_reasons_to_filter)
]
```

```python
# Ok, now we have filtered_detentions_df which is EVERY SINGLE ROW wwhere 'Stay Release Reason' is found in release_reasons_to_filter list.
# This means we've got multiple rows probably showing a person inside one faciltiy or another, though their final stay status is the same.
# So now we want to do what we did earlier: isolate the UNIQUE stays. So one person's entire, single stay results in a release of some sort.

unique_filtered_bookings = filtered_detentions_df[['Unique Identifier', 'Stay Book In Date Time']].drop_duplicates()
```
